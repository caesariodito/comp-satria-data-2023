```python
import os
from PIL import Image, ImageFile

# ImageFile.LOAD_TRUNCATED_IMAGES = True

root_folder_path = './data-modified/Data Train for BDC 2023 - Penyisihan/data/'

errors = []

for folder_path, subfolders, filenames in os.walk(root_folder_path):
    counter = 1
    for filename in filenames:
        file_path = os.path.join(folder_path, filename)
        print(file_path)
        if os.path.isfile(file_path):
            filename, file_extension = os.path.splitext(file_path)
            # new_file_name = '{:06d}.jpg'.format(counter)
            # new_file_path = os.path.join(folder_path, new_file_name)
            if file_extension.lower() in ['.webp', '.jpeg', '.png']:
                try:
                    with Image.open(file_path) as image:
                        if image.mode == 'RGBA' or image.mode == 'P':
                            image = image.convert('RGB')
                        print('saved file:', filename+file_extension)
                        image.save(filename+file_extension)
                    os.remove(file_path)
                except:
                    errors.append(file_path)
                    os.remove(file_path)
            else:
                # os.rename(file_path, new_file_path)
                pass
            counter += 1
```
